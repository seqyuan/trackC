````python
import pyBigWig

import trackc as tc

AML_1360_H3K27ac = pyBigWig.open(
    "../../trackc_data/tutorials/bigwig/GSM4604189_H3K27ac.bw"
)
````

````python
ten = tc.tenon(figsize=(8, 1))
ten.add(pos="bottom", height=1, hspace=0.1)
ten.add(pos="bottom", height=1, hspace=0.2)
ten.add(pos="bottom", height=0.5, hspace=0.2)


regions = ["chr8:127000000-129200000", "chr14:96500000-99300000"]
tc.pl.bw_track(
    AML_1360_H3K27ac,
    ten.axs(0),
    regions=regions,
    maxrange=20,
    label="H3K27ac",
    binsize=10000,
    color="tab:blue",
)
tc.pl.bw_track(
    AML_1360_H3K27ac,
    ten.axs(1),
    regions=regions,
    maxrange=5,
    label="H3K27ac",
    binsize=10000,
    invert_y=True,
    ax_on=True,
)
tc.pl.multi_scale_track(
    ten.axs(2), regions=regions, scale_adjust="Mb", intervals=1, tick_fontsize=10
)

# tc..savefig('trackc_bigwig.pdf')
````
