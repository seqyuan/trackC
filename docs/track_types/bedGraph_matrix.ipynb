## data
HFF_MicroC.mcool: https://files.osf.io/v1/resources/hjcxg/providers/osfstorage/5f73756d9e9a3d03ec6e3922?action=download&direct&version=1

```python
df = pd.read_table("./trackc_data/tutorials/bedGraphMatrix/HFF_MicroC.chr2is.txt")
coolfile = '/Users/yuanzan/Downloads/HFF_MicroC.mcool::resolutions/10000'

regions = "chr2:10500000-13200000"
HFF = tc.tl.extractCisContact(clr=coolfile, region=regions, extend=150)

ten = tc.tenon(figsize=(7, 1))
ten.add(pos="bottom", height=1.5, hspace=0.05)
ten.add(pos="bottom", height=0.5, hspace=0.07)
ten.add(pos="bottom", height=1, hspace=0.07)

tc.pl.mapC(
    ax=ten.axs(0),
    mat=HFF,
    label="HFF",
    map_type="rec",
    height=150,
)

tc.pl.bdgmat_track(ax=ten.axs(1), bed=df, regions=regions, cmap='RdBu_r')
tc.pl.bdgmat_track(ax=ten.axs(2), bed=df, regions=regions, style='line')
tc.pl.scale_track(ax=ten.axs(2), region=regions, scale_adjust="Mb", tick_pos="bottom", ratio2ax=1)

#tc.savefig('outfile.pdf')
```
