conf-hicmap_1.yml

```yaml
trackc:
  - ax: t1
    height: 2
    track_type: hicmap
    track_para:
      mat:
        method: extractContactRegions
        clr: /path/GSM4417639_SK-N-DZ_C6BC81F2_b38d5.mcool::/resolutions/5120000
        row_regions: 
          - chr6
          - chr8
      mapC:
        cmap: plasma
```
