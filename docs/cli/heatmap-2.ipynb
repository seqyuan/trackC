# two chrom trans interaction
## CLI mode

`conf-hicmap_2.yml`

```yaml
trackc:
  - ax: t1
    height: 4
    track_type: hicmap
    track_para:
      mat:
        method: extractContactRegions
        clr: /path/GSM4417639_SK-N-DZ_C6BC81F2_b38d5.mcool::/resolutions/5120000
        row_regions: chr6
        col_regions: chr8
      mapC:
        cmap: Blues
        map_type: squ
        symmetric: True
        maxrange: 1000
        aspect: 1
```

```shell
trackc cli conf-hicmap_2.yml -s 4,1
```



## API mode

````python
import cooler
import trackc as tc

ten = tc.tenon(figsize=(4, 1))
ten.add(pos="bottom", height=4)


clr = "./GSM4417639_SK-N-DZ_C6BC81F2_b38d5.mcool::/resolutions/5120000"

mat = tc.tl.extractContactRegions(
    clr=clr,
    row_regions=['chr6'],
    col_regions=['chr8']
)

tc.pl.mapC(ax=ten.axs(0), 
           mat=mat.cmat, 
           map_type="squ", 
           cmap='Blues', 
           symmetric=True,
           maxrange= 1000,
           aspect=1,
          )
#tc.savefig('chr6_chr8_trans.pdf')
````
