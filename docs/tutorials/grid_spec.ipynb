## trackc.tenon

```python
ten = tc.tenon(figsize=(5, 0.7))
ten.add(pos='bottom', height=4)
ten.add(pos='bottom', height=0.5, hspace=0.1)
ten.add(pos='top', height=1, hspace=0.1)
ten.axs(0).plot([0,1])
ten.show()
```
